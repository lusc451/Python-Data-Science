```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
# Definindo o título do gráfico
plt.title('Valores do combustível em US$')


# Criando o primeiro gráfico
plt.plot(gas['Year'], gas['Australia'], 'r.-', label = 'Australia') # Deixando a linha vermelha e pontilhada através do parametro 'r.--'
plt.plot(gas['Year'], gas['Italy'], 'y.-', label= 'Italy')
plt.plot(gas['Year'], gas['USA'], 'b.-', label= 'USA')
plt.plot(gas['Year'], gas['Japan'], 'g.-', label= 'Japan')

# Deixando a contagem de anos visualmente mais agradável
plt.xticks(gas['Year'][::2]) # Contagem de anos pulando a cada 2 anos

# Definindo legendas para os eixos X e Y
plt.xlabel('Anos')
plt.ylabel('Valores em US$')

# Mostrando a legenda
plt.legend()

# Mostrando o gráfico
plt.show()
```
