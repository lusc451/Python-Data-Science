```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
# gas
```

```python
# Definindo o ttulo do grfico
plt.title('Valores do combustvel em US$')


# Criando o primeiro grfico
plt.plot(gas['Year'], gas['Australia'], label = 'Australia') # Definindo a legenda do grfico atravs do parmetro label
plt.plot(gas['Year'], gas['Italy'], label= 'Italy')
plt.plot(gas['Year'], gas['USA'], label= 'USA')
plt.plot(gas['Year'], gas['Japan'], label= 'Japan')

# Deixando a contagem de anos visualmente mais agradvel
plt.xticks(gas['Year'][::2]) # Contagem de anos pulando a cada 2 anos

# Mostrando a legenda
plt.legend()

# Mostrando o grfico
plt.show()
```
