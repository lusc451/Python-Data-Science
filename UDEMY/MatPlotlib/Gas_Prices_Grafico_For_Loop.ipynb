```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
# Definindo o ttulo do grfico
plt.title('Valores do combustvel em US$')

#Criando uma lista com pases do DataSet e montando o grfico atravs de um For Loop
list_country = ['Australia', 'USA', 'Canada', 'Italy', 'France', 'Mexico', 'Germany']

for country in gas:
    if country in list_country:
        plt.plot(gas['Year'], gas[country], label= [country], marker= '.')

# Deixando a contagem de anos visualmente mais agradvel
plt.xticks(gas['Year'][::2]) # Contagem de anos pulando a cada 2 anos

# Definindo legendas para os eixos X e Y
plt.xlabel('Anos')
plt.ylabel('Valores em US$')

# Mostrando a legenda
plt.legend()

# Salvando o grfico em vrios formatos **SEMPRE SALVAR O GRFICO ANTES DA EXIBIO**
plt.savefig('country.png')
plt.savefig('country.pdf')
plt.savefig('country.jpeg')

# Mostrando o grfico
plt.show()
```
